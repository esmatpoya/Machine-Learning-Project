```python
X = wine_dataset.drop(['quality'], axis= 1)
```

```python
X
```

```python
Y = wine_dataset['quality'].apply(lambda y_value:1 if y_value >= 7  else 0 )
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X_train.shape, X_test.shape,Y_train.shape, Y_test.shape)
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
train_model_prediction = model.predict(X_train)
train_accuracy_prediction = accuracy_score(train_model_prediction, Y_train)

train_accuracy_prediction
```

```python
test_model_prediction = model.predict(X_test)

test_acuracy_prediction = accuracy_score(test_model_prediction, Y_test)
test_acuracy_prediction
```
